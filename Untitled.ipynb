# Starting Codes

```python
import jovian
import random as rn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("googleplaystore.csv")
df.sample(10)
```

```python
df.info();
```

```python
df.Category.describe()
```

```python
df.drop([10472], inplace= True)
```

```python
df.Installs = df.Installs.apply(lambda x: str(x).replace("+",""))
df.Installs = df.Installs.apply(lambda x: str(x).replace(",",""))
df.Installs = df.Installs.apply(lambda x: int(x))
df.Installs.dtype
```

```python
clr = ['Red','Orange','Skyblue','Blue','Aqua','Green']
sel_clr = rn.choice(clr)
```

# Data

```python
df
```

```python
df.shape
```

# Data Engineering

```python
df.isnull().sum()
```

# Top 10 Download Apps code

```python
top_app_instl = df.groupby('App').Installs.sum().sort_values(ascending=False).head(10)
top_app_instl
```

# Top 10 Downloaded App

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 Downloaded Apps')
plt.xticks(rotation=90)
sns.barplot(top_app_instl.index,top_app_instl.values);
```

# Highest Pricing Apps Code

```python
df.Price = df.Price.apply(lambda x: str(x).replace("$",""))
df.Price = df.Price.apply(lambda x: str(x).replace(",",""))
df.Price = df.Price.apply(lambda x: float(x))
df.Price.dtype
```

```python
top_price =  df.groupby('App').Price.sum().sort_values(ascending=False).head(20)
```

# Highest Pricing Apps 

```python
plt.figure(figsize=(10,6))
plt.title("TOP DOWNLOADED APP")
plt.xticks(rotation= 90)
sns.scatterplot(top_price.index,top_price.values)
plt.plot(top_price.index,top_price.values,'g-o')
sns.barplot(top_price.index,top_price.values);
```

# What is the rating of each Category


```python
count= 1
plt.figure(figsize=(25,35))
for i in np.unique(df.Category.values):
    plt.subplot(11,3,count);
    plt.hist(df[df.Category == i].Rating.values,color=sel_clr);
    plt.title(i)
    plt.xlabel('Rating')
    plt.ylabel('Count')
    count += 1
plt.tight_layout()
    
```

# Most Category Reviewed

```python
df.Reviews = df.Reviews.apply(lambda x: int(x))
```

```python
Reviewed = df.groupby('Category').Reviews.sum().sort_values(ascending=False).head(10)
Reviewed
```

```python
plt.figure(figsize=(15,6))
plt.title('Top 10 Downloaded Apps')
plt.xticks(rotation=90)
sns.barplot(Reviewed.index,Reviewed);
```
